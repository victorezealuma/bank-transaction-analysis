## Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv(r"/content/drive/MyDrive/bank_transactions.csv")

df
```

```python
df = df.rename(columns={'TransactionAmount (INR)': 'TransactionAmount(NGN)'})


df
```

```python
df['TransactionTime'] = df['TransactionTime'].apply(lambda x: str(x))

df['TransactionTime'].apply(lambda x: x[0:2] + ':' + x[2:4] + ':'+ x[4:6])

df['TransactionTime'] = df['TransactionTime'].apply(lambda x: x[0:2] + ':' + x[2:4] + ':'+ x[4:6])

df
```

```python
data = {'CustLocation': ['MUMBAI', 'NEW DELHI', 'BANGALORE', 'GURGAON','DELHI','PUNE','NAVI MUMBAI']}

df['CustLocation'] = df['CustLocation'].replace('MUMBAI', 'LAGOS ISLAND')
df['CustLocation'] = df['CustLocation'].replace('NEW DELHI', 'IKEJA')
df['CustLocation'] = df['CustLocation'].replace('BANGALORE', 'YABA')
df['CustLocation'] = df['CustLocation'].replace('GURGAON', 'IKORODU')
df['CustLocation'] = df['CustLocation'].replace('DELHI', 'LEKKI')
df['CustLocation'] = df['CustLocation'].replace('PUNE', 'ABULE-EGBA')
df['CustLocation'] = df['CustLocation'].replace('NAVI MUMBAI', 'MUSHIN')

df
```
